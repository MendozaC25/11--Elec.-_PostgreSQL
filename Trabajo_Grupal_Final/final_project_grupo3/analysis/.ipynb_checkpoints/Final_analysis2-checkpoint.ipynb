## 3. Análisis de querys

```python
# Consulta JOIN Avanzado: Obtener registros combinados de felicidad y migración
df_advanced_join = advanced_join_happiness_migration(engine)
print("Primeros 5 registros de la consulta JOIN Avanzado:")
display(df_advanced_join.head())
```

```python
# Obtener la tabla de relación entre GDP per Cápita y Migración Neta
df_relation_gdp = relation_gdp_migration(engine)

# Gráfico: Relación entre GDP per Cápita y Migración Neta (Promedio 2015-2019)
plt.figure(figsize=(10,6))
sns.scatterplot(x='avg_gdp_per_capita', y='avg_net_migration', data=df_relation_gdp, s=100, color="teal")
plt.title("Relación entre GDP per Cápita y Migración Neta (2015-2019)")
plt.xlabel("GDP per Cápita Promedio")
plt.ylabel("Migración Neta Promedio")
plt.tight_layout()
plt.show()
```

```python
# Consulta agrupada: Promedio de Felicidad por Año (con filtros HAVING)
df_grouped = group_by_year_with_having(engine)
print("Primeros 5 registros de la Consulta Agrupada (Promedio de Felicidad por Año):")
display(df_grouped.head())
```

```python
# Gráfico: Promedio de Felicidad por Año obtenido de la consulta agrupada
plt.figure(figsize=(10,6))
ax = sns.barplot(x='year', y='promedio_felicidad', data=df_grouped, palette="coolwarm")
plt.title("Promedio de Felicidad por Año (Consulta Agrupada)")
plt.xlabel("Año")
plt.ylabel("Promedio de Felicidad")

# Agregar etiquetas con los valores sobre cada barra
for p in ax.patches:
    height = p.get_height()
    plt.text(p.get_x() + p.get_width() / 2, 
             height + 0.1, 
             f'{height:.2f}', 
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Consulta: Clasificación Condicional de Felicidad y Migración
df_conditional = conditional_aggregation(engine)
print("Primeros 5 registros de la Consulta de Clasificación Condicional:")
display(df_conditional.head())
```

```python
if 'categoria_felicidad' in df_conditional.columns:
    plt.figure(figsize=(10,6))
    ax = sns.countplot(x='categoria_felicidad', data=df_conditional, palette="magma")
    plt.title("Distribución de Categorías de Felicidad")
    plt.xlabel("Categoría")
    plt.ylabel("Frecuencia")
    
    # Agregar etiquetas con los valores sobre cada barra
    for p in ax.patches:
        height = p.get_height()
        plt.text(p.get_x() + p.get_width() / 2, 
                 height + 0.1, 
                 f'{int(height)}', 
                 ha='center', va='bottom')
    
    plt.tight_layout()
    plt.show()
else:
    print("No se encontró la columna 'categoria_felicidad' para graficar.")
```

```python
# Consulta: Ranking Top5 de Felicidad por Año
df_top5_rank = top5_happiness_rank(engine)
print("Ranking Top 5 de Felicidad por Año:")
display(df_top5_rank.head())
```

```python
# Filtrar para el año 2019 (si la columna 'year' existe)
if 'year' in df_top5_rank.columns:
    df_top5_2019 = df_top5_rank[df_top5_rank['year'] == 2019]
    plt.figure(figsize=(10,6))
    ax = sns.barplot(x='happiness_score', y='country', data=df_top5_2019, palette="viridis")
    plt.title("Top 5 Países con Mejor Felicidad - Año 2019")
    plt.xlabel("Puntaje de Felicidad")
    plt.ylabel("País")
    
    # Agregar etiquetas con los valores sobre cada barra
    for p in ax.patches:
        width = p.get_width()
        plt.text(width + 0.01, 
                 p.get_y() + p.get_height() / 2, 
                 f'{width:.2f}', 
                 va='center')
    
    plt.tight_layout()
    plt.show()
else:
    print("La columna 'year' no está presente en los datos de ranking.")
```

```python
# Consulta: Evolución de la Felicidad por País a lo largo de los años
df_evolution = happiness_evolution(engine)
print("Primeros 5 registros de la Evolución de la Felicidad:")
display(df_evolution.head())
```

```python
# Cambia el valor de 'pais' según se desee analizar
pais = 'Afghanistan'
df_evolucion_pais = df_evolution[df_evolution['country'] == pais]

plt.figure(figsize=(10,6))
ax = sns.lineplot(x='year', y='happiness_score', data=df_evolucion_pais, marker='o', linewidth=2)
plt.title(f"Evolución de la Felicidad en {pais}")
plt.xlabel("Año")
plt.ylabel("Puntaje de Felicidad")
plt.grid(True)

# Agregar etiquetas con los valores en cada punto
for index, row in df_evolucion_pais.iterrows():
    plt.text(row['year'], row['happiness_score'] + 0.1, f"{row['happiness_score']:.2f}", 
             horizontalalignment='center')

plt.tight_layout()
plt.show()
```

```python
# Consulta: Ranking de Países por Promedio de Felicidad (2015-2019)
df_avg_happiness = ranking_avg_happiness(engine)
print("Ranking de Países por Promedio de Felicidad (2015-2019):")
display(df_avg_happiness.head())

# Ordenar el DataFrame de forma descendente
df_avg_happiness_sorted = df_avg_happiness.sort_values('avg_happiness_score', ascending=False)
# Tomar los 5 primeros registros
df_avg_top5 = df_avg_happiness_sorted.head(5)
```

```python
# Gráfico: Top 5 Países en promedio de felicidad con etiquetas de valor
plt.figure(figsize=(10,6))
ax = sns.barplot(x='avg_happiness_score', y='country', data=df_avg_top5, palette="Spectral")
plt.title("Ranking de Países por Promedio de Felicidad (Top 5, 2015-2019)")
plt.xlabel("Promedio de Felicidad")
plt.ylabel("País")

# Agregar etiquetas con el valor sobre cada barra
for p in ax.patches:
    width = p.get_width()
    # Posición horizontal de la etiqueta
    plt.text(width + 0.01, p.get_y() + p.get_height()/2, f'{width:.2f}', va='center')

plt.tight_layout()
plt.show()
```

### Análisis 
Para el gráfico es un bar plot que muestra los 5 países con el mayor promedio de felicidad durante el período de 2015 a 2019, en el eje X representa el promedio de felicidad (happiness score) de cada país. Y en el eje Y, los países. Este gráfico nos muestra los cinco países con los mayores promedios de felicidad entre 2015 y 2019, entre los cuales en promedio son bastante parecidos. Entre los países que descatcan, Denmark encabeza la lista con un promedio de felicidad de 7.55, seguido muy de cerca por Norway y Finland con 7.54 cada uno. Iceland y Switzerland también destacan con un promedio de 7.51. Este gráfico nos pareció que es fundamental para el trabajo porque identifica rápidamente los líderes en felicidad, también nos permite analizar cómo sus políticas migratorias pueden influir en la felicidad de sus habitantes, además, facilita la comparación de estos países con otros en términos de felicidad y migración. 

```python
# Consulta: Ranking de Países con Migración Positiva (2015-2019)
df_positive_migration = ranking_positive_migration(engine)
print("Ranking de Países con Migración Positiva (2015-2019):")
display(df_positive_migration.head())

# Ordenar el DataFrame descendientemente según la columna 'avg_net_migration' y tomar los 5 primeros
df_positive_migration_sorted = df_positive_migration.sort_values('avg_net_migration', ascending=False)
df_positive_top5 = df_positive_migration_sorted.head(5)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='avg_net_migration', y='country', data=df_positive_top5, palette="Blues_d")
plt.title("Ranking de Países con Migración Positiva (Top 5, 2015-2019)")
plt.xlabel("Migración Neta")
plt.ylabel("País")

# Agregar etiquetas con los valores sobre cada barra
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 0.01, 
             p.get_y() + p.get_height() / 2, 
             f'{width:.2f}', 
             va='center')

plt.tight_layout()
plt.show()
```

### Análisis
Este cuadro muestra los cinco países con la mayor migración neta promedio de 2015 a 2019. Estados Unidos lidera con una migración neta promedio de 1,281,618, seguido de Alemania y Canadá. Colombia y Reino Unido también están en el top 5. Este gráfico es importante ya que nos ayuda a identificar rápidamente a los países que reciben más migrantes, analizar los factores que los hacen atractivos. Hemos llegado a la conclusión que estos países pueden ser más atractivos debido a que tienen economías más fuertes, un buen acceso a servicios de salud y educación de calidad, estabilidad política y social y por ende un nivel de felicidad mayor de las personas que viven en esos países. 


```python
# Consulta: Ranking de Países con Menor Migración Promedio (Valor Absoluto) 2015-2019
df_lowest_migration = ranking_lowest_migration(engine)
print("Ranking de Países con Menor Migración Promedio (Valor Absoluto) (2015-2019):")
display(df_lowest_migration.head())

# Ordenar el DataFrame (se asume que la menor migración es la que tiene valores absolutos menores)
df_lowest_migration_sorted = df_lowest_migration.sort_values('avg_abs_net_migration', key=lambda x: abs(x))
df_lowest_top5 = df_lowest_migration_sorted.head(5)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(x='avg_abs_net_migration', y='country', data=df_lowest_top5, palette="Greens_d")
plt.title("Ranking de Países con Menor Migración Promedio (Valor Absoluto) (Top 5, 2015-2019)")
plt.xlabel("Migración Neta")
plt.ylabel("País")

# Agregar etiquetas con los valores sobre cada barra
for p in ax.patches:
    width = p.get_width()
    plt.text(width + (0.01 if width >= 0 else -0.01), 
             p.get_y() + p.get_height() / 2, 
             f'{width:.2f}', 
             va='center', 
             ha='left' if width >= 0 else 'right')

plt.tight_layout()
plt.show()
```

### Análisis 

El gráfico muestra los cinco países con la menor migración neta promedio (en valor absoluto) durante el período 2015-2019, lo cual contraste el gráfico que anteriormente analizámos. Bhutan encabeza la lista con 513 personas por año, seguido por Madagascar, Suriname, North Macedonia y Azerbaijan. Los valores bajos de migración neta promedio sugieren que no son países tan atractivos. Bhutan, con la menor migración neta promedio, podría estar influenciado por sus políticas migratorias o debido a factores como la falta de oportunidades económicas, sistemas de salud y educación deficientes, altos niveles de corrupción y pobreza. Estas condiciones desfavorables pueden desincentivar tanto la inmigración como la emigración, resultando en una baja migración neta promedio.

```python
# Consulta: Relación entre Percepción de Corrupción y Felicidad (Promedio 2015-2019)
df_relation_corr = relation_corruption_happiness(engine)
print("Relación entre Percepción de Corrupción y Felicidad (Promedio 2015-2019):")
display(df_relation_corr.head())
```

### Análisis 

Lo observando en el cuadro que realizamos es la relación entre la percepción de corrupción y los niveles de felicidad promedio en diferentes países (top 5), se puede notar un patrón interesante. Los datos muestran que a medida que la percepción de corrupción aumenta, los niveles de felicidad tienden a disminuir. Por ejemplo, Bosnia y Herzegovina, con una percepción de corrupción muy baja (0.002), tiene un nivel de felicidad promedio de 5.162, mientras que Trinidad y Tobago, con una mayor percepción de corrupción (0.011), tiene un nivel de felicidad promedio de 6.168.

Esto sugiere que los países con una baja percepción de corrupción pueden no ser necesariamente más felices. Sin embargo, otros factores como la calidad de vida, la economía, la estabilidad política y la confianza en las instituciones también juegan un papel crucial en determinar la felicidad de la población. Es probable que los países con alta corrupción enfrenten mayores desafíos sociales y económicos que impactan negativamente en la felicidad de sus ciudadanos. Este análisis nos pareció importante ya que nos ayudaría a combatir la corrupción para mejorar el bienestar general de las países

```python
plt.figure(figsize=(10,6))
sns.regplot(
    x='avg_corruption_perception',
    y='avg_happiness_score',
    data=df_relation_corr,
    scatter=True,
    scatter_kws={'s': 100, 'color': "orange"},
    line_kws={'color': "blue"}
)
plt.title("Relación entre Percepción de Corrupción y Felicidad (2015-2019)")
plt.xlabel("Índice de Corrupción")
plt.ylabel("Promedio de Felicidad")
plt.tight_layout()
plt.show()
```

### Análisis 

El gráfico muestra la relación entre la percepción de corrupción y los niveles de felicidad promedio de diferentes países de 2015 a 2019. El eje X representa la percepción de corrupción, mientras que el eje Y indica el promedio de felicidad. Los puntos en el gráfico revelan una correlación negativa: a medida que aumenta la percepción de corrupción, los niveles de felicidad tienden a disminuir. Esto sugiere que en los países con mayor percepción de corrupción, las personas tienden a ser menos felices. De igual forma, es importante considerar que también como lo analizamos en el trabajo, hay factores como la calidad del sistema de salud, educación, estabilidad económica y niveles de pobreza, que también pueden influyen en la percepción general de felicidad.
